```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Load dataset
df = pd.read_csv("ipl2022.csv")
numeric_cols = [
    'first_ings_score',
    'first_ings_wkts',
    'second_ings_score',
    'second_ings_wkts',
    'margin',
    'highscore'
]

# PDF & CDF plots
plt.figure(figsize=(14, 10))
for i, col in enumerate(numeric_cols, 1):
    plt.subplot(2, 3, i)
    
    # Sort values for CDF
    sorted_data = np.sort(df[col])
    cdf = np.arange(1, len(sorted_data) + 1) / len(sorted_data)
    
    # Plot PDF using seaborn (KDE)
    sns.kdeplot(df[col], color='blue', fill=True, label='PDF')
    
    # Plot CDF
    plt.plot(sorted_data, cdf, color='red', label='CDF', linewidth=2)
    
    plt.title(f"PDF & CDF of {col}")
    plt.xlabel(col)
    plt.ylabel("Probability")
    plt.legend()

plt.tight_layout()
plt.show()
```

```python
import pandas as pd


numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns

pearson_skewness = {}

for col in numeric_cols:
    mean_val = df[col].mean()
    median_val = df[col].median()
    std_val = df[col].std()
    
    if std_val != 0:  # avoid division by zero
        skew_val = 3 * (mean_val - median_val) / std_val
    else:
        skew_val = None
    
    pearson_skewness[col] = skew_val


pearson_skewness = pd.Series(pearson_skewness, name="Pearson_Skewness")
print(pearson_skewness)
```

```python
import pandas as pd
import numpy as np
from scipy.stats import boxcox
import matplotlib.pyplot as plt


cols_to_transform = ['margin', 'highscore']


transformed_data = {}
lambdas = {}

for col in cols_to_transform:
   
    data = df[col].values
    if np.any(data <= 0):
        data = data + 1  
    
    
    transformed, lam = boxcox(data)
    transformed_data[col] = transformed
    lambdas[col] = lam
    
  
    plt.figure(figsize=(10,4))
    plt.subplot(1,2,1)
    plt.hist(df[col], bins=20, color='skyblue', edgecolor='black')
    plt.title(f"Original {col}\nSkewness: {df[col].skew():.3f}")
    
    plt.subplot(1,2,2)
    plt.hist(transformed, bins=20, color='salmon', edgecolor='black')
    plt.title(f"Box-Cox {col}\nÎ»={lam:.3f}, Skewness: {pd.Series(transformed).skew():.3f}")
    plt.show()

print("Optimal Box-Cox Î» values:")
print(lambdas)
```

```python
import pandas as pd
from scipy.stats import kurtosis

numeric_df = df.select_dtypes(include=['int64', 'float64'])

def classify_kurtosis(value):
    if abs(value - 3) < 0.1:   # allow a small tolerance
        return "Mesokurtic (Normal-like)"
    elif value > 3:
        return "Leptokurtic (Heavy tails, peaked)"
    else:
        return "Platykurtic (Light tails, flat)"

# Compute kurtosis (Pearson definition)
kurtosis_values = numeric_df.apply(lambda x: kurtosis(x, fisher=False))

# Print values with type
print("--- Kurtosis Classification ---")
for col, val in kurtosis_values.items():
    print(f"{col}: {val:.3f} â†’ {classify_kurtosis(val)}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import kurtosis

numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns

plt.figure(figsize=(15, 10))

for i, col in enumerate(numeric_cols, 1):
    plt.subplot(3, 3, i)  
    
   
    sns.histplot(df[col], kde=True, bins=30, color="skyblue", edgecolor="black")
    
  
    k_value = kurtosis(df[col].dropna(), fisher=False)  # Pearson definition
    
  
    if abs(k_value - 3) < 0.1:
        k_type = "Mesokurtic"
    elif k_value > 3:
        k_type = "Leptokurtic"
    else:
        k_type = "Platykurtic"
    
    plt.title(f"{col}\nKurtosis = {k_value:.2f} â†’ {k_type}")

plt.tight_layout()
plt.show()
```
