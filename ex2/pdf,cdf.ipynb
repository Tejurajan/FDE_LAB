```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import binom

# Parameters
n = 3       # number of tosses
p = 0.5     # probability of heads

x = np.arange(n + 1)
pmf = binom.pmf(x, n, p)
cdf = binom.cdf(x, n, p)

# Plot setup
plt.figure(figsize=(12, 5))

# PMF
plt.subplot(1, 2, 1)
plt.stem(x, pmf, linefmt='blue', markerfmt='bo', basefmt='k-')
plt.title('PMF: P(X = k)')
plt.xlabel('Number of Heads (k)')
plt.ylabel('Probability')
plt.xticks(x)
plt.grid(True)

# CDF
plt.subplot(1, 2, 2)
plt.step(x, cdf, where='post', color='orange', marker='o')
plt.title('CDF: F(k) = P(X â‰¤ k)')
plt.xlabel('Number of Heads (k)')
plt.ylabel('Cumulative Probability')
plt.xticks(x)
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import binom

n = 100      # number of coin tosses
p = 0.5      # fair coin

# Define the range of X (number of heads)
x = np.arange(0, n + 1)

# Calculate PMF and CDF
pmf = binom.pmf(x, n, p)
cdf = binom.cdf(x, n, p)

# Plotting
plt.figure(figsize=(14, 6))

# PMF Plot
plt.subplot(1, 2, 1)
plt.plot(x, pmf, color='blue')
plt.title('PMF: Probability Mass Function ')
plt.xlabel('Number of Heads (X)')
plt.ylabel('P(X = x)')
plt.grid(True)

# CDF Plot
plt.subplot(1, 2, 2)
plt.plot(x, cdf, color='orange')
plt.title('CDF: Cumulative Distribution Function ')
plt.xlabel('Number of Heads (X)')
plt.ylabel('P(X â‰¤ x)')
plt.grid(True)

plt.tight_layout()
plt.show()
```
