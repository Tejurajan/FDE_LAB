```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform


x1 = uniform.rvs(loc=0, scale=1, size=100000, random_state=1)
x2 = uniform.rvs(loc=0, scale=1, size=100000, random_state=2)


z = x1 + x2


plt.hist(z, bins=50, density=True, color='skyblue', edgecolor='black')
plt.title('Distribution of z = x1 + x2 ')
plt.xlabel('z')
plt.ylabel('Density')
plt.grid(True)
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform


x1 = uniform.rvs(loc=0, scale=1, size=100000, random_state=1)
x2 = uniform.rvs(loc=0, scale=2, size=100000, random_state=2)


z = x1 + x2


plt.hist(z, bins=50, density=True, color='skyblue', edgecolor='black')
plt.title('Distribution of z = x1 + x2 ')
plt.xlabel('z')
plt.ylabel('Density')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform


N = 100_000


x1, x2, x3, x4, x5, x6, x7, x8, x9, x10 = [
    uniform.rvs(loc=0, scale=1, size=N, random_state=seed)
    for seed in range(1, 11)
]

z = x1 + x2 + x3 + x4 + x5 + x6 + x7 + x8 + x9 + x10


plt.hist(z, bins=50, density=True, color='skyblue', edgecolor='black')
plt.title('Distribution of z = x1 + ... + x10')
plt.xlabel('z')
plt.ylabel('Density')
plt.grid(True)
plt.show()
```
