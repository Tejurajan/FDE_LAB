```python
from scipy.stats import norm

# Standard normal distribution: mean=0, std=1
mean = 0
std = 1

# P(0)
p0 = norm.cdf(0, loc=mean, scale=std)
print(f"P(x <= 0): {p0}")

# P(-1 < x <= 1)
p1 = norm.cdf(1, loc=mean, scale=std) - norm.cdf(-1, loc=mean, scale=std)
print(f"P(-1 < x <= 1): {p1}")

# P(-2 < x <= 2)
p2 = norm.cdf(2, loc=mean, scale=std) - norm.cdf(-2, loc=mean, scale=std)
print(f"P(-2 < x <= 2): {p2}")

# P(-3 < x <= 3)
p3 = norm.cdf(3, loc=mean, scale=std) - norm.cdf(-3, loc=mean, scale=std)
print(f"P(-3 < x <= 3): {p3}")
```

```python
from scipy.stats import norm


mean = 6
std = 12

# P(0)
p0 = norm.cdf(mean, loc=mean, scale=std)
print(f"P(x <= 0): {p0}")

# P(-1 < x <= 1)
p1 = norm.cdf(mean - std, loc=mean, scale=std) - norm.cdf(mean + std, loc=mean, scale=std)
print(f"P(-1 < x <= 1): {p1}")

# P(-2 < x <= 2)
p2 = norm.cdf(mean + 2*std, loc=mean, scale=std) - norm.cdf(mean - 2*std, loc=mean, scale=std)
print(f"P(-2 < x <= 2): {p2}")

# P(-3 < x <= 3)
p3 = norm.cdf(mean + 3*std, loc=mean, scale=std) - norm.cdf(mean - 3*std, loc=mean, scale=std)
print(f"P(-3 < x <= 3): {p3}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters
mean = 6
std = 12

# Values for x
x = np.linspace(mean - 4*std, mean + 4*std, 1000)
y = norm.pdf(x, loc=mean, scale=std)

# Probabilities
p0 = norm.cdf(0, loc=mean, scale=std)
p1 = norm.cdf(mean + std, loc=mean, scale=std) - norm.cdf(mean - std, loc=mean, scale=std)
p2 = norm.cdf(mean + 2*std, loc=mean, scale=std) - norm.cdf(mean - 2*std, loc=mean, scale=std)
p3 = norm.cdf(mean + 3*std, loc=mean, scale=std) - norm.cdf(mean - 3*std, loc=mean, scale=std)

# Print probabilities
print(f"P(x <= 0): {p0:.4f}")
print(f"P(mean ± 1σ): {p1:.4f}")
print(f"P(mean ± 2σ): {p2:.4f}")
print(f"P(mean ± 3σ): {p3:.4f}")

# Plot
plt.figure(figsize=(12, 6))
plt.plot(x, y, color='black', label='Normal Distribution')

# Shade ±1σ
plt.fill_between(x, y, where=((x > mean - std) & (x < mean + std)), color='skyblue', alpha=0.5, label='±1σ')

# Shade ±2σ
plt.fill_between(x, y, where=((x > mean - 2*std) & (x < mean + 2*std)), color='lightgreen', alpha=0.4, label='±2σ')

# Shade ±3σ
plt.fill_between(x, y, where=((x > mean - 3*std) & (x < mean + 3*std)), color='salmon', alpha=0.3, label='±3σ')

# Line for x = 0
plt.axvline(0, color='red', linestyle='--', label='x = 0')

# Formatting
plt.title("Normal Distribution with Mean = 6 and Std Dev = 12")
plt.xlabel("x")
plt.ylabel("Probability Density")
plt.legend()
plt.grid(True)
plt.show()
```
