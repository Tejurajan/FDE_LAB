```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform, kstest

# Step 1: Generate uniform data between 5 and 10
data = np.random.uniform(5, 10, size=1000)

# Step 2: Define the uniform distribution
a = 5       # lower bound
b = 10      # upper bound
dist = uniform(loc=a, scale=b - a)  # Uniform(a, b)

# Step 3: Prepare x values for PDF and CDF
x = np.linspace(4, 11, 500)
pdf = dist.pdf(x)
cdf = dist.cdf(x)

# Step 4: Plot PDF, CDF, and Histogram
plt.figure(figsize=(12, 5))

# PDF & Histogram
plt.subplot(1, 2, 1)
plt.hist(data, bins=20, density=True, alpha=0.5, color='skyblue', edgecolor='black', label='Histogram')
plt.plot(x, pdf, 'r-', label='PDF', linewidth=2)
plt.title('Uniform Distribution PDF (a=5, b=10)')
plt.xlabel('x')
plt.ylabel('Density')
plt.legend()
plt.grid(True)

# CDF
plt.subplot(1, 2, 2)
plt.plot(x, cdf, 'g-', label='CDF', linewidth=2)
plt.title('Uniform Distribution CDF (a=5, b=10)')
plt.xlabel('x')
plt.ylabel('Cumulative Probability')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()

# Step 5: Uniformity Test (Kolmogorov-Smirnov)
data_scaled = (data - a) / (b - a)  # scale to [0, 1]
ks_stat, p_value = kstest(data_scaled, 'uniform')

print(f"\nKolmogorov-Smirnov test statistic: {ks_stat:.4f}")
print(f"P-value: {p_value:.4f}")
if p_value > 0.05:
    print(" Data is likely uniformly distributed (fail to reject H0)")
else:
    print(" Data is not uniformly distributed (reject H0)")
```
