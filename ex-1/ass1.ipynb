```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 1000)

# Function and its derivative
y = np.exp(-x)
y_deriv = -np.exp(-x)

# Plotting
fig, axs = plt.subplots(1, 2, figsize=(12, 4))

axs[0].plot(x, y, color='blue', label='f(x) = e^(-x)')
axs[0].set_title('Function: f(x) = e^(-x)')
axs[0].axhline(0, color='gray', linestyle='--')
axs[0].axvline(0, color='gray', linestyle='--')
axs[0].grid(True)
axs[0].legend()

axs[1].plot(x, y_deriv, color='red', label="f'(x) = -e^(-x)")
axs[1].set_title("Derivative: f'(x) = -e^(-x)")
axs[1].axhline(0, color='gray', linestyle='--')
axs[1].axvline(0, color='gray', linestyle='--')
axs[1].grid(True)
axs[1].legend()

plt.tight_layout()
plt.show()
```

```python
# Re-import after code execution reset
import numpy as np
import matplotlib.pyplot as plt

# Define x values
x = np.linspace(-10, 10, 1000)

# Mod (absolute) function
def mod(x):
    return np.abs(x)

# Derivative of mod function
def mod_derivative(x):
    return np.where(x > 0, 1, np.where(x < 0, -1, 0))  # Approximating derivative at x=0 as 0

# Plotting
fig, axs = plt.subplots(1, 2, figsize=(12, 4))

# Plot mod function
axs[0].plot(x, mod(x), color='blue', label='|x|')
axs[0].set_title('Modulus Function: f(x) = |x|')
axs[0].axhline(0, color='gray', linestyle='--', linewidth=0.5)
axs[0].axvline(0, color='gray', linestyle='--', linewidth=0.5)
axs[0].grid(True)
axs[0].legend()

# Plot derivative
axs[1].plot(x, mod_derivative(x), color='red', label="f'(x)")
axs[1].set_title("Derivative of |x|")
axs[1].axhline(0, color='gray', linestyle='--', linewidth=0.5)
axs[1].axvline(0, color='gray', linestyle='--', linewidth=0.5)
axs[1].grid(True)
axs[1].legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define x values
x = np.linspace(-10, 10, 500)

# Linear function and its derivative
f = x
f_derivative = np.ones_like(x)  # Derivative of x is 1

# Plotting
fig, axs = plt.subplots(1, 2, figsize=(10, 4))

# Function plot
axs[0].plot(x, f, color='blue', label='f(x) = x')
axs[0].set_title('Linear Function: f(x) = x')
axs[0].axhline(0, color='gray', linestyle='--')
axs[0].axvline(0, color='gray', linestyle='--')
axs[0].legend()
axs[0].grid(True)

# Derivative plot
axs[1].plot(x, f_derivative, color='red', label="f'(x) = 1")
axs[1].set_title("Derivative: f'(x) = 1")
axs[1].axhline(0, color='gray', linestyle='--')
axs[1].axvline(0, color='gray', linestyle='--')
axs[1].legend()
axs[1].grid(True)

plt.tight_layout()
plt.show()
```

```python
def relu(x):
    return np.maximum(0, x)

def relu_derivative(x):
    return np.where(x > 0, 1, 0)

# Leaky ReLU function and its derivative
def leaky_relu(x, alpha=0.01):
    return np.where(x > 0, x, alpha * x)

def leaky_relu_derivative(x, alpha=0.01):
    return np.where(x > 0, 1, alpha)
activation_functions = [
    ("ReLU", relu, relu_derivative),
    ("Leaky ReLU", leaky_relu, leaky_relu_derivative),
   
]

fig, axs = plt.subplots(len(activation_functions), 2, figsize=(14, 18))

for i, (name, func, deriv) in enumerate(activation_functions):
    axs[i, 0].plot(x, func(x), color='blue', label=f"{name} Function")
    axs[i, 0].set_title(f"{name} Activation Function")
    axs[i, 0].axhline(0, color='gray', linestyle='--', linewidth=0.5)
    axs[i, 0].grid(True)
    axs[i, 0].legend()

    axs[i, 1].plot(x, deriv(x), color='red', label=f"{name} Derivative")
    axs[i, 1].set_title(f"{name} Derivative")
    axs[i, 1].axhline(0, color='gray', linestyle='--', linewidth=0.5)
    axs[i, 1].grid(True)
    axs[i, 1].legend()

plt.tight_layout()
plt.show()
```
