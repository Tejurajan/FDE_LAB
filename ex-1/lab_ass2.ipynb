```python
import pandas as pd
df = pd.read_csv("fitness_tracker_dataset.csv") 
df.head()
```

```python
from sklearn.preprocessing import StandardScaler

features_to_normalize = ['Age', 'Height (cm)', 'Weight (kg)', 'Daily Steps']

scaler = StandardScaler()
df_zscore = df.copy()
df_zscore[features_to_normalize] = scaler.fit_transform(df[features_to_normalize])

df_zscore.head()
```

```python
zscore_scaler = StandardScaler()
df_zscore = df.copy()
df_zscore[features_to_normalize] = zscore_scaler.fit_transform(df[features_to_normalize])
print("\nZ-Score Normalized")
print(df_zscore[features_to_normalize].agg(['min', 'max']))
```

```python
import matplotlib.pyplot as plt

features = ['Age', 'Height (cm)', 'Weight (kg)', 'Daily Steps']

df[features].hist(bins=10, figsize=(12, 8), color='skyblue', edgecolor='black')
plt.suptitle("Histograms of Feature Ranges", fontsize=16)
plt.tight_layout()
plt.show()
```
